```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
taxi = pd.read_csv('taxi_monthly.csv', header =0)
weather = pd.read_csv('weather_monthly.txt', sep=',', header =0)
```

```python
weather.head()
```

```python
weatax = pd.merge(taxi, weather, how='inner', on=['Date'])
```

```python
weatax.head()
```

```python
nodate = list(weather.columns.values)
del nodate[0]
```

```python
print(nodate)
```

```python
for item in nodate:
    print(item)
```

```python
x = list(weatax['Date'])
for item in nodate:
    item = str(item)
    yweath = list(weatax[item])
    ytax = list(weatax['count'])
    fig = plt.figure()
    ax1 = fig.add_subplot(111)
    ax1.plot(yweath,'b-')
    ax1.set_ylabel(item)
    ax2 = ax1.twinx()
    ax2.plot(ytax,'r-')
    ax2.set_ylabel('Number of taxi trips')
    ax1.set_xlabel('Year and Month (yyyymm)')
    plt.title('Number of taxi trips vs %s'%(item))
```
